```python
X = [0, 1, 2, 3]
y = [0, 0, 1, 1]

w = 0.0
b = -0.2
lr = 0.1
epochs = 10


def activation(z):
    return 1 if z >= 0 else 0


for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):

        z = w * X[i] + b

        y_pred = activation(z)

        error = y[i] - y_pred

        w = w + lr * error * X[i]


        print(f"X={X[i]}, Target={y[i]}, Pred={y_pred}, Error={error}")
        print(f"Updated w={w}, b={b}")

b = b + lr * error

print("\nFinal Results:")
for i in range(len(X)):
    z = w * X[i] + b
    print(f"Input {X[i]} -> Output {activation(z)}")
```
